```python
#Create a bar chart showing sales by product category
#Make a line chart showing monthly temperature changes
#Build a pie chart showing expense distribution
#dd titles, labels to make charts professional
#Combine data loading, cleaning, analysis and visualization
#Create a complete report with both numbers and charts

!pip install reportlab
import pandas as pd
import matplotlib.pyplot as plt
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.pdfbase.cidfonts import UnicodeCIDFont
from reportlab.pdfbase import pdfmetrics
import os

pdfmetrics.registerFont(UnicodeCIDFont("HeiseiKakuGo-W5"))

# Create the directory if it doesn't exist
os.makedirs('/mnt/data/', exist_ok=True)

# 1. CREATE SYNTHETIC DATASETS

sales_data = pd.DataFrame({
    "Category": ["Electronics", "Furniture", "Clothing", "Food", "Books"],
    "Sales": [52000, 34000, 27000, 45000, 15000]
})

temperature_data = pd.DataFrame({
    "Month": pd.date_range(start="2024-01-01", periods=12, freq="M"),
    "Temperature": [2, 5, 9, 14, 18, 22, 25, 24, 20, 14, 8, 3]
})

expense_data = pd.DataFrame({
    "Expense": ["Rent", "Salaries", "Utilities", "Marketing", "Misc"],
    "Amount": [3000, 7000, 1200, 2500, 800]
})


# 2. CLEANING

sales_data.dropna(inplace=True)
temperature_data.dropna(inplace=True)
expense_data.dropna(inplace=True)


# 3. ANALYSIS

summary_sales = sales_data.describe().to_string()
summary_temp = temperature_data.describe().to_string()
summary_expense = expense_data.describe().to_string()


# 4. VISUALIZATION

# Bar chart
plt.figure(figsize=(6,4))
plt.bar(sales_data["Category"], sales_data["Sales"])
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
bar_path = "/mnt/data/bar_chart.png"
plt.savefig(bar_path)

# Line chart
plt.figure(figsize=(6,4))
plt.plot(temperature_data["Month"], temperature_data["Temperature"])
plt.title("Monthly Temperature Changes")
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.tight_layout()
line_path = "/mnt/data/line_chart.png"
plt.savefig(line_path)

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(expense_data["Amount"], labels=expense_data["Expense"], autopct="%1.1f%%")
plt.title("Expense Distribution")
plt.tight_layout()
pie_path = "/mnt/data/pie_chart.png"
plt.savefig(pie_path)


```

```python
# 5. BUILD  REPORT

report_path = "/mnt/data/Complete_Report.pdf"
styles = getSampleStyleSheet()
styleN = styles["Normal"]
styleH = styles["Heading1"]

doc = SimpleDocTemplate(report_path, pagesize=letter)
elements = []

elements.append(Paragraph("Complete Data Analysis Report", styleH))
elements.append(Spacer(1, 12))

# Sales summary
elements.append(Paragraph("<b>Sales Summary</b>", styleN))
elements.append(Paragraph(summary_sales.replace("\n", "<br/>"), styleN))
elements.append(Spacer(1, 12))
elements.append(Image(bar_path, width=5*inch, height=3*inch))
elements.append(Spacer(1, 24))

# Temperature summary
elements.append(Paragraph("<b>Temperature Summary</b>", styleN))
elements.append(Paragraph(summary_temp.replace("\n", "<br/>"), styleN))
elements.append(Spacer(1, 12))
elements.append(Image(line_path, width=5*inch, height=3*inch))
elements.append(Spacer(1, 24))

# Expense summary
elements.append(Paragraph("<b>Expense Summary</b>", styleN))
elements.append(Paragraph(summary_expense.replace("\n", "<br/>"), styleN))
elements.append(Spacer(1, 12))
elements.append(Image(pie_path, width=4.5*inch, height=4.5*inch))
elements.append(Spacer(1, 24))

doc.build(elements)

report_path
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# SYNTHETIC STUDENT PERFORMANCE DATA
# -----------------------------
np.random.seed(42)

students = pd.DataFrame({
    "Student": [f"Student {i}" for i in range(1, 21)],
    "Math": np.random.randint(50, 100, 20),
    "Science": np.random.randint(45, 95, 20),
    "English": np.random.randint(55, 100, 20),
})

# Calculate averages
students["Average"] = students[["Math", "Science", "English"]].mean(axis=1)

# -----------------------------
# ANALYSIS
# -----------------------------
subject_means = students[["Math", "Science", "English"]].mean()
monthly_progress = pd.DataFrame({
    "Month": pd.date_range("2024-01-01", periods=6, freq="M"),
    "Average_Score": np.linspace(60, 85, 6) + np.random.randint(-3, 3, 6)
})

# -----------------------------
# BAR CHART  Average scores by subject
# -----------------------------
plt.figure(figsize=(6,4))
plt.bar(subject_means.index, subject_means.values)
plt.title("Average Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.tight_layout()
bar_path = "/mnt/data/subject_averages.png"
plt.savefig(bar_path)

# -----------------------------
# LINE CHART  Monthly progress
# -----------------------------
plt.figure(figsize=(6,4))
plt.plot(monthly_progress["Month"], monthly_progress["Average_Score"])
plt.title("Student Average Score Progress Over Months")
plt.xlabel("Month")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.tight_layout()
line_path = "/mnt/data/monthly_progress.png"
plt.savefig(line_path)

# Return paths for user download
bar_path, line_path
```
