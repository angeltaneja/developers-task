```python
#3. Load the dataset & display first few rows
import pandas as pd

df = pd.read_csv("iris.csv", header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Display first few rows
print(df.head())
```

```python
#4. Check basic information
# Number of rows and columns
print(df.shape)

# Data types and non-null counts
print(df.info())

# Statistical summary
print(df.describe())
```

```python
#5. Find and handle missing values

# Check for missing values
missing_values = df.isnull().sum()

# Display missing values count per column
print("Missing values per column:")
print(missing_values)

# Since there are no missing values, no further handling is needed for this dataset.
```

```python
# 6. Calculate average, highest, lowest values
# Average of each numeric column
print(df.mean(numeric_only=True))

# Maximum values
print(df.max(numeric_only=True))

# Minimum values
print(df.min(numeric_only=True))
```

```python
# project
#Analyze a Simple Sales Dataset to find: total sales, best-selling product, and create a basic report
```

```python
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam', 'Microphone', 'Headphones', 'Speaker', 'External HDD', 'SSD']
quantities = [10, 25, 15, 8, 30, 20, 12, 5, 7, 18]
prices = [1200.00, 25.00, 75.00, 300.00, 45.00, 60.00, 90.00, 150.00, 80.00, 110.00]

sales_data = {
    'Product': products,
    'Quantity': quantities,
    'Price': prices
}

sales_df = pd.DataFrame(sales_data)
print("Sales DataFrame created:")
print(sales_df.head())
# total sales
sales_df['Sales'] = sales_df['Quantity'] * sales_df['Price']
total_sales = sales_df['Sales'].sum()

print("Total Sales:", total_sales)
# best selling product
product_sales = sales_df.groupby('Product')['Sales'].sum()
best_selling_product = product_sales.idxmax()
max_sales = product_sales.max()

print(f"Best-Selling Product: {best_selling_product}")
print(f"Sales Value: {max_sales}")
#visulaize product sales
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
product_sales.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales per Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
