# Task
Generate diverse `seaborn` visualizations in Python, including box plots, correlation heatmaps, and various chart layouts like subplots and dashboards, while demonstrating custom styling and clarifying the static nature of `seaborn` plots.

## Generate Sample Data

### Subtask:
Create a sample pandas DataFrame with diverse numerical and categorical data types to serve as the basis for all subsequent visualization tasks.


**Reasoning**:
To fulfill the subtask, I will first import the pandas library, then create a dictionary with diverse data types as specified, and finally convert it into a pandas DataFrame named `df`.



```python
import pandas as pd
import numpy as np

data = {
    'ID': range(1, 11),
    'Value_A': np.random.randint(10, 100, 10),
    'Value_B': np.random.uniform(5.0, 50.0, 10),
    'Category_X': np.random.choice(['Red', 'Blue', 'Green'], 10),
    'Category_Y': np.random.choice(['Alpha', 'Beta'], 10),
    'Date': pd.to_datetime(pd.date_range('2023-01-01', periods=10, freq='D')),
    'Boolean_Flag': np.random.choice([True, False], 10)
}

df = pd.DataFrame(data)
print("Sample DataFrame created successfully:")
print(df.head())
```

## Create Box Plots

### Subtask:
Generate box plots using `seaborn` to visualize the distribution and potential outliers of numerical data across different categories within the sample dataset.


**Reasoning**:
I will import the necessary libraries, then create a box plot visualizing the distribution of 'Value_A' across 'Category_X' using `seaborn`, and set appropriate labels and a title.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Category_X', y='Value_A', data=df)
plt.xlabel('Category X')
plt.ylabel('Value A')
plt.title('Box Plot of Value A by Category X')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("Box plot of 'Value_A' by 'Category_X' generated successfully.")
```

**Reasoning**:
To further address the subtask of generating diverse box plots, I will create another box plot. This time, I will visualize the distribution of 'Value_B' across 'Category_Y' using `seaborn`, and ensure appropriate labels and a title are set.



```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category_Y', y='Value_B', data=df)
plt.xlabel('Category Y')
plt.ylabel('Value B')
plt.title('Box Plot of Value B by Category Y')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("Box plot of 'Value_B' by 'Category_Y' generated successfully.")
```

## Build Correlation Heatmaps

### Subtask:
Calculate the correlation matrix for the numerical features in the sample data and visualize it as a heatmap using `seaborn.heatmap`, including annotations for clarity.


**Reasoning**:
To calculate the correlation matrix, I will first select the numerical columns from the DataFrame `df`. Then, I will compute the correlation matrix for these columns and visualize it using a seaborn heatmap with annotations and a descriptive title, as requested in the instructions.



```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
print("Correlation heatmap of numerical features generated successfully.")
```

## Design Professional-looking Charts with Custom Styles

### Subtask:
Demonstrate how to apply `seaborn`'s built-in styles (e.g., 'darkgrid', 'whitegrid') and custom color palettes to create aesthetically pleasing and professional-looking charts.


**Reasoning**:
To fulfill the subtask, I will apply a seaborn style, create a scatter plot using the specified columns and apply a custom color palette based on 'Category_X', add appropriate labels and a title, display the plot, and finally reset the seaborn style to default.



```python
sns.set_style('darkgrid')

plt.figure(figsize=(10, 7))
sns.scatterplot(x='Value_A', y='Value_B', hue='Category_X', data=df, palette='viridis', s=100)
plt.xlabel('Value A')
plt.ylabel('Value B')
plt.title('Scatter Plot of Value A vs. Value B by Category X with Darkgrid Style')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("Scatter plot with custom style and palette generated successfully.")
sns.reset_defaults()
print("Seaborn style reset to default.")
```

## Create Multiple Subplots in a Single Figure

### Subtask:
Utilize `matplotlib.pyplot.subplots` in conjunction with `seaborn` plotting functions to arrange multiple distinct plots within a single figure, demonstrating how to control layout and sharing axes.


**Reasoning**:
To address the subtask, I will create a figure with two subplots, generating a seaborn box plot on the first subplot and a seaborn scatter plot on the second, applying titles to each and ensuring proper layout.



```python
fig, axs = plt.subplots(1, 2, figsize=(18, 7))

sns.boxplot(x='Category_X', y='Value_A', data=df, ax=axs[0])
axs[0].set_title('Distribution of Value A by Category X')
axs[0].set_xlabel('Category X')
axs[0].set_ylabel('Value A')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

sns.scatterplot(x='Value_A', y='Value_B', data=df, ax=axs[1], hue='Category_X', palette='viridis', s=100)
axs[1].set_title('Scatter Plot of Value A vs. Value B')
axs[1].set_xlabel('Value A')
axs[1].set_ylabel('Value B')
axs[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
print("Multiple subplots with box plot and scatter plot generated successfully.")
```

Generate a `seaborn` plot and provide a note explaining that `seaborn` primarily creates static plots. For true interactivity with hover information, libraries such as Plotly, Bokeh, or Altair are typically used in conjunction or as alternatives.

## Combine Multiple Charts into a Dashboard Layout

### Subtask:
Arrange several different `seaborn` visualizations (e.g., box plot, histogram, scatter plot) into a grid-like 'dashboard' layout within a single figure using `matplotlib.pyplot.GridSpec` or `subplots`.


**Reasoning**:
To fulfill the subtask, I will create a 2x2 grid of subplots using `matplotlib.pyplot.subplots`, and then populate each subplot with a different seaborn visualization: a box plot, a histogram, a scatter plot, and a count plot, ensuring appropriate titles and labels for each, and finally use `plt.tight_layout()` and `plt.show()` to display the dashboard.



```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Subplot 1: Box plot of Value A by Category X
sns.boxplot(x='Category_X', y='Value_A', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Value A by Category X')
axs[0, 0].set_xlabel('Category X')
axs[0, 0].set_ylabel('Value A')
axs[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 2: Histogram of Value B
sns.histplot(df['Value_B'], kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Distribution of Value B')
axs[0, 1].set_xlabel('Value B')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 3: Scatter plot of Value A vs. Value B by Category Y
sns.scatterplot(x='Value_A', y='Value_B', hue='Category_Y', data=df, ax=axs[1, 0], palette='magma', s=100)
axs[1, 0].set_title('Scatter Plot of Value A vs. Value B by Category Y')
axs[1, 0].set_xlabel('Value A')
axs[1, 0].set_ylabel('Value B')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Subplot 4: Count plot of Category X
sns.countplot(x='Category_X', data=df, ax=axs[1, 1], palette='viridis')
axs[1, 1].set_title('Count of Category X')
axs[1, 1].set_xlabel('Category X')
axs[1, 1].set_ylabel('Count')
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
print("Dashboard layout with multiple seaborn visualizations generated successfully.")
```

**Reasoning**:
The previous code block generated a `FutureWarning` because `palette` was used with `sns.countplot` without assigning `hue`. To fix this, I will explicitly set `hue='Category_X'` and `legend=False` in the `countplot` call, as suggested by the warning.



```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Subplot 1: Box plot of Value A by Category X
sns.boxplot(x='Category_X', y='Value_A', data=df, ax=axs[0, 0])
axs[0, 0].set_title('Distribution of Value A by Category X')
axs[0, 0].set_xlabel('Category X')
axs[0, 0].set_ylabel('Value A')
axs[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 2: Histogram of Value B
sns.histplot(df['Value_B'], kde=True, ax=axs[0, 1])
axs[0, 1].set_title('Distribution of Value B')
axs[0, 1].set_xlabel('Value B')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 3: Scatter plot of Value A vs. Value B by Category Y
sns.scatterplot(x='Value_A', y='Value_B', hue='Category_Y', data=df, ax=axs[1, 0], palette='magma', s=100)
axs[1, 0].set_title('Scatter Plot of Value A vs. Value B by Category Y')
axs[1, 0].set_xlabel('Value A')
axs[1, 0].set_ylabel('Value B')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Subplot 4: Count plot of Category X
sns.countplot(x='Category_X', hue='Category_X', data=df, ax=axs[1, 1], palette='viridis', legend=False)
axs[1, 1].set_title('Count of Category X')
axs[1, 1].set_xlabel('Category X')
axs[1, 1].set_ylabel('Count')
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
print("Dashboard layout with multiple seaborn visualizations generated successfully.")
```

## Generate Sample Sales Data

### Subtask:
Create a sample pandas DataFrame representing sales data, including columns for sales trends (e.g., date, revenue), customer segmentation (e.g., customer ID, segment, region), and product performance (e.g., product ID, category, quantity, price). This data will be the foundation for all subsequent visualization tasks.


**Reasoning**:
To fulfill the subtask, I will import the necessary libraries, create a dictionary with diverse sales data including sales trends, customer segmentation, and product performance, convert it into a pandas DataFrame named `df_sales`, and then display its first few rows.



```python
import pandas as pd
import numpy as np

# Generate sample data
num_entries = 100

data = {
    'Date': pd.to_datetime(pd.date_range('2023-01-01', periods=num_entries, freq='D')),
    'OrderID': range(1001, 1001 + num_entries),
    'CustomerID': np.random.randint(1, 50, num_entries),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], num_entries),
    'ProductCategory': np.random.choice(['Electronics', 'Clothing', 'Home Goods', 'Books'], num_entries),
    'ProductID': np.random.randint(100, 200, num_entries),
    'Quantity': np.random.randint(1, 10, num_entries),
    'PricePerUnit': np.round(np.random.uniform(10.0, 500.0, num_entries), 2),
    'CustomerSegment': np.random.choice(['New', 'Returning', 'Wholesale', 'Loyal'], num_entries)
}

df_sales = pd.DataFrame(data)
df_sales['Revenue'] = df_sales['Quantity'] * df_sales['PricePerUnit']

print("Sample Sales DataFrame created successfully:")
print(df_sales.head())
```

## Seaborn Basic Plots and Styling (Day 1)

### Subtask:
Generate basic `seaborn` plots (e.g., line plot for sales trends, bar plot for product sales) using the sample sales data. Apply custom colors and styles to demonstrate aesthetic customization.


### Reasoning:
To address the subtask, I will first import the `seaborn` and `matplotlib.pyplot` libraries. Next, I will set a `seaborn` style for aesthetic customization. I will then prepare the data for sales trends by grouping `df_sales` by `Date` and summing `Revenue`, followed by creating a line plot. Similarly, I will prepare the data for product sales by grouping `df_sales` by `ProductCategory` and summing `Revenue`, and then create a bar plot with a custom color palette. Finally, I will reset the `seaborn` style to its default settings to ensure subsequent plots are not affected by this customization.

**Reasoning**:
I will generate the Python code to import libraries, set a seaborn style, create a line plot for daily sales trends, create a bar plot for total revenue by product category, and finally reset the seaborn style as per the detailed instructions.



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
print("Seaborn style set to 'whitegrid'.")

# 3. Create a line plot for sales trends
daily_revenue = df_sales.groupby('Date')['Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Revenue', data=daily_revenue, color='skyblue')
plt.title('Daily Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print("Line plot of daily sales trends generated successfully.")

# 4. Create a bar plot for product sales
category_revenue = df_sales.groupby('ProductCategory')['Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='ProductCategory', y='Revenue', data=category_revenue, palette='viridis')
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("Bar plot of total revenue by product category generated successfully.")

# 5. Reset the seaborn style
sns.reset_defaults()
print("Seaborn style reset to default.")
```

**Reasoning**:
The previous `sns.barplot` call produced a `FutureWarning` because `palette` was used without `hue`. To address this, I will explicitly set `hue='ProductCategory'` and `legend=False` in the `sns.barplot` function, as suggested by the warning, to ensure compatibility and suppress the warning.



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
print("Seaborn style set to 'whitegrid'.")

# 3. Create a line plot for sales trends
daily_revenue = df_sales.groupby('Date')['Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Revenue', data=daily_revenue, color='skyblue')
plt.title('Daily Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print("Line plot of daily sales trends generated successfully.")

# 4. Create a bar plot for product sales
category_revenue = df_sales.groupby('ProductCategory')['Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='ProductCategory', y='Revenue', hue='ProductCategory', data=category_revenue, palette='viridis', legend=False)
plt.title('Total Revenue by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print("Bar plot of total revenue by product category generated successfully.")

# 5. Reset the seaborn style
sns.reset_defaults()
print("Seaborn style reset to default.")
```
