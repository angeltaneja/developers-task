```python
#1. Program that gives different messages based on time of day
hour = int(input("Enter the current hour (0–23): "))

if 5 <= hour < 12:
    print("Good morning!")
elif 12 <= hour < 17:
    print("Good afternoon!")
elif 17 <= hour < 21:
    print("Good evening!")
else:
    print("Good night!")
```

```python
#2. Number guessing game using if-else
secret_number = 7
guess = int(input("Guess the number (1–10): "))

if guess == secret_number:
    print("Correct! You guessed the number!")
elif guess < secret_number:
    print("Too low!")
else:
    print("Too high!")
```

```python
#4. Function that calculates area of a rectangle
def rectangle_area(width, height):
    return width * height

w = float(input("Enter width: "))
h = float(input("Enter height: "))

print("Area =", rectangle_area(w, h))
```

```python
#5. Simple calculator (add, subtract, multiply, divide)
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))

if operator == "+":
    print("Result:", num1 + num2)
elif operator == "-":
    print("Result:", num1 - num2)
elif operator == "*":
    print("Result:", num1 * num2)
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero.")
    else:
        print("Result:", num1 / num2)
else:
    print("Invalid operator.")
```

```python
# 6 Real-life scenarios
score = int(input("Enter score (0–100): "))

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

```python
# project
#Build a Simple Student Grade Calculator that takes marks and returns grade (A, B, C, etc.) with encouraging messages for each grade
marks = float(input("Enter the student's marks (0–100): "))

if marks >= 90:
    grade = "A"
    message = "Excellent work! Keep aiming high."
elif marks >= 80:
    grade = "B"
    message = "Great job! You're doing really well."
elif marks >= 70:
    grade = "C"
    message = "Good effort! Keep practicing and you'll get even better."
elif marks >= 60:
    grade = "D"
    message = "You're improving. Stay focused and keep trying."
else:
    grade = "F"
    message = "Don’t give up. Every step is progress, and you can still succeed."

print("Grade:", grade)
print("Message:", message)
```
