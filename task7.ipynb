## Analyze Data Distribution

### Subtask:
Generate histograms for key numerical variables to visualize their distributions and perform statistical tests (e.g., Shapiro-Wilk) to check for normal distribution.


**Reasoning**:
The subtask requires visualizing data distributions using histograms and performing statistical tests for normality. I will import necessary libraries (`matplotlib.pyplot` for plotting and `scipy.stats.shapiro` for the normality test) and then generate the code to create histograms for 'Sales', 'Units', and 'Price' columns, followed by performing the Shapiro-Wilk test on each of these columns and printing their results.



```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Create a figure with subplots for histograms
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram for Sales
axes[0].hist(df['Sales'], bins=5, edgecolor='black')
axes[0].set_title('Distribution of Sales')
axes[0].set_xlabel('Sales')
axes[0].set_ylabel('Frequency')

# Histogram for Units
axes[1].hist(df['Units'], bins=5, edgecolor='black')
axes[1].set_title('Distribution of Units')
axes[1].set_xlabel('Units')
axes[1].set_ylabel('Frequency')

# Histogram for Price
axes[2].hist(df['Price'], bins=5, edgecolor='black')
axes[2].set_title('Distribution of Price')
axes[2].set_xlabel('Price')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Perform Shapiro-Wilk test for Sales
shapiro_sales = shapiro(df['Sales'])
print(f"\nShapiro-Wilk Test for Sales: Statistic={shapiro_sales.statistic:.3f}, p-value={shapiro_sales.pvalue:.3f}")

# Perform Shapiro-Wilk test for Units
shapiro_units = shapiro(df['Units'])
print(f"Shapiro-Wilk Test for Units: Statistic={shapiro_units.statistic:.3f}, p-value={shapiro_units.pvalue:.3f}")

# Perform Shapiro-Wilk test for Price
shapiro_price = shapiro(df['Price'])
print(f"Shapiro-Wilk Test for Price: Statistic={shapiro_price.statistic:.3f}, p-value={shapiro_price.pvalue:.3f}")
```

## Perform Hypothesis Tests

### Subtask:
Formulate and test at least three hypotheses related to the sales data using appropriate statistical tests (ANOVA and t-tests), and interpret the p-values.


**Reasoning**:
As instructed, I will first import the necessary statistical functions (`f_oneway` and `ttest_ind`) from `scipy.stats` to prepare for the hypothesis tests.



```python
from scipy.stats import f_oneway, ttest_ind
print("f_oneway and ttest_ind imported from scipy.stats")
```

**Reasoning**:
Now, I will perform the first hypothesis test: an ANOVA test to determine if sales differ significantly across different regions. This involves grouping sales data by region, applying the `f_oneway` function, and then printing the F-statistic and p-value for interpretation.



```python
print("\nHypothesis 1: Test if sales differ significantly across different regions")
regions = df['Region'].unique()
sales_by_region = [df['Sales'][df['Region'] == r] for r in regions]

f_statistic_region, p_value_region = f_oneway(*sales_by_region)
print(f"ANOVA Test for Sales across Regions: F-statistic={f_statistic_region:.3f}, p-value={p_value_region:.3f}")

if p_value_region < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in sales across regions.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in sales across regions.")
```

**Reasoning**:
Next, I will perform the second hypothesis test, which involves an ANOVA test to determine if sales differ significantly across different products. I will group sales data by product, apply the `f_oneway` function from `scipy.stats`, and then print the F-statistic and p-value for interpretation, along with a conclusion based on a significance level of 0.05.



```python
print("\nHypothesis 2: Test if sales differ significantly across different products")
products = df['Product'].unique()
sales_by_product = [df['Sales'][df['Product'] == p] for p in products]

f_statistic_product, p_value_product = f_oneway(*sales_by_product)
print(f"ANOVA Test for Sales across Products: F-statistic={f_statistic_product:.3f}, p-value={p_value_product:.3f}")

if p_value_product < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in sales across products.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in sales across products.")
```

**Reasoning**:
Now I will perform the third hypothesis test, a t-test, to determine if there is a significant difference in sales between the 'East' and 'West' regions. I will extract the sales data for these two regions, apply the `ttest_ind` function from `scipy.stats`, and then print the t-statistic and p-value for interpretation, along with a conclusion based on a significance level of 0.05.



```python
print("\nHypothesis 3: Test if there is a significant difference in sales between the 'East' and 'West' regions.")
sales_east = df['Sales'][df['Region'] == 'East']
sales_west = df['Sales'][df['Region'] == 'West']

t_statistic_ew, p_value_ew = ttest_ind(sales_east, sales_west)
print(f"Independent t-test for Sales between East and West: t-statistic={t_statistic_ew:.3f}, p-value={p_value_ew:.3f}")

if p_value_ew < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in sales between East and West regions.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in sales between East and West regions.")
```

## Calculate Confidence Intervals

### Subtask:
Calculate 95% confidence intervals for key metrics, such as average sales, to determine the margin of error and the range within which the true population parameter is likely to fall.


**Reasoning**:
To calculate the 95% confidence intervals for the average sales as per the subtask, I will import the `t` class, then calculate the mean, standard error of the mean (SEM), and degrees of freedom for the 'Sales' column. Finally, I will compute and print the confidence interval.



```python
from scipy.stats import t

# Calculate the mean of the 'Sales' column
mean_sales = df['Sales'].mean()

# Calculate the standard error of the mean (SEM) for the 'Sales' column
sem_sales = df['Sales'].sem()

# Determine the degrees of freedom
degrees_of_freedom = len(df['Sales']) - 1

# Calculate the 95% confidence interval for the mean of 'Sales'
confidence_interval_sales = t.interval(confidence=0.95, df=degrees_of_freedom, loc=mean_sales, scale=sem_sales)

print(f"\nAverage Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval for Average Sales: {confidence_interval_sales}")
print(f"Lower Bound: {confidence_interval_sales[0]:.2f}")
print(f"Upper Bound: {confidence_interval_sales[1]:.2f}")
```
