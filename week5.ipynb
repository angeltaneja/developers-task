```python
import pandas as pd
import io

# Sample sales data string (assuming this was available in the original environment)
csv_data = """Date,Product,Quantity,Price,Customer_ID,Region,Total_Sales
2024-01-01,Phone,7,37300,CUST001,East,261100
2024-01-02,Headphones,4,15406,CUST002,North,61624
2024-01-03,Phone,2,21746,CUST003,West,43492
2024-01-04,Headphones,1,30895,CUST004,East,30895
2024-01-05,Laptop,8,39835,CUST005,North,318680
2024-01-06,Phone,3,25000,CUST006,South,75000
2024-01-07,Headphones,6,18000,CUST007,West,108000
2024-01-08,Laptop,1,45000,CUST008,East,45000
2024-01-09,Phone,5,30000,CUST009,North,150000
2024-01-10,Headphones,2,12000,CUST010,South,24000
"""

# Load sales data from the existing csv_data string
sales_df = pd.read_csv(io.StringIO(csv_data), parse_dates=["Date"])

##Extract Year, Month, Day from Date
sales_df["Year"] = sales_df["Date"].dt.year
sales_df["Month"] = sales_df["Date"].dt.month
sales_df["Day"] = sales_df["Date"].dt.day

print(sales_df.head())
```

## Filter Data Using Multiple Conditions

### Subtask:
Filter the `sales_df` DataFrame using multiple conditions, demonstrating both AND and OR operations.


**Reasoning**:
Filter the `sales_df` DataFrame using multiple conditions, first applying an AND condition to find high sales in the 'East' region, and then an OR condition to find sales of 'Phone' products or in the 'North' region, finally displaying the head of both filtered DataFrames to verify the results.



```python
##Filter Data Using Multiple Conditions (AND / OR)

# 1. Filter for 'Region' is 'East' AND 'Total_Sales' are greater than 100000
east_high_sales_df = sales_df[
    (sales_df["Region"] == "East") & (sales_df["Total_Sales"] > 100000)
]

# 2. Filter for 'Product' is 'Phone' OR 'Region' is 'North'
phone_or_north_sales_df = sales_df[
    (sales_df["Product"] == "Phone") | (sales_df["Region"] == "North")
]

# 3. Print the head of both filtered DataFrames
print("East Region and High Sales (Total_Sales > 100000):")
print(east_high_sales_df.head())
print("\nPhone Product or North Region Sales:")
print(phone_or_north_sales_df.head())
```

## Clean and Transform Text Data

### Subtask:
Identify a suitable text column (e.g., 'Product' or 'Region') and apply string methods like converting to lowercase, stripping whitespace, or replacing specific characters to demonstrate text data cleaning.


**Reasoning**:
To clean and transform the text data as per the subtask, I will convert the 'Product' column to lowercase and strip any leading/trailing whitespace, then display the unique values to confirm the cleaning.



```python
## Clean and Transform Text Data

# 1. Convert the 'Product' column to lowercase
sales_df["Product"] = sales_df["Product"].str.lower()

# 2. Remove any leading or trailing whitespace from the 'Product' column
sales_df["Product"] = sales_df["Product"].str.strip()

# 3. Display the unique values of the cleaned 'Product' column
print("Unique values in cleaned 'Product' column:")
print(sales_df["Product"].unique())
```

## Create and Merge Customer Data

### Subtask:
Generate a sample `customer_df` DataFrame with 'Customer_ID' and 'Customer_Segment' columns, then merge it with the `sales_df` DataFrame based on 'Customer_ID'.


**Reasoning**:
To fulfill the subtask, I will create a sample `customer_df` DataFrame with 'Customer_ID' and 'Customer_Segment' columns, ensuring some IDs overlap with `sales_df`. Then, I will perform a left merge of `sales_df` with `customer_df` on 'Customer_ID' to add the customer segment information. Finally, I will display the head of the merged `sales_df` to verify the new column.



```python
## Create and Merge Customer Data

# 1. Create a sample customer_df DataFrame
customer_data = {
    'Customer_ID': ['CUST001', 'CUST002', 'CUST003', 'CUST004', 'CUST005', 'CUST011', 'CUST012'],
    'Customer_Segment': ['Premium', 'Regular', 'Premium', 'Regular', 'Premium', 'New', 'New']
}
customer_df = pd.DataFrame(customer_data)

print("Customer DataFrame head:")
print(customer_df.head())

# 2. Merge sales_df with customer_df based on 'Customer_ID'
sales_df = pd.merge(sales_df, customer_df, on='Customer_ID', how='left')

# 3. Display the first few rows of the merged DataFrame
print("\nMerged Sales DataFrame head with Customer Segment:")
print(sales_df.head())
```

## Create Pivot Tables

### Subtask:
Generate pivot tables to summarize sales data by different categories, such as 'Region', 'Product', and 'Customer_Segment', showing aggregate statistics like total sales.


**Reasoning**:
To fulfill the subtask's requirements, I will create three separate pivot tables. Each pivot table will summarize the 'Total_Sales' using different categories: 'Region', 'Product', and 'Customer_Segment'. I will then print each pivot table to display the aggregated sales data, confirming the successful generation of these summaries.

**Reasoning**:
To create the pivot tables as per the subtask instructions, I will use the `pivot_table` function to group and aggregate 'Total_Sales' by 'Region', 'Product', and 'Customer_Segment' and then print each resulting table.



```python
## Create Pivot Tables

# 1. Pivot table for total sales by Region
pivot_region_sales = pd.pivot_table(
    sales_df, values='Total_Sales', index='Region', aggfunc='sum'
)
print("\nTotal Sales by Region:")
print(pivot_region_sales)

# 2. Pivot table for total sales by Product
pivot_product_sales = pd.pivot_table(
    sales_df, values='Total_Sales', index='Product', aggfunc='sum'
)
print("\nTotal Sales by Product:")
print(pivot_product_sales)

# 3. Pivot table for total sales by Customer_Segment
pivot_segment_sales = pd.pivot_table(
    sales_df, values='Total_Sales', index='Customer_Segment', aggfunc='sum'
)
print("\nTotal Sales by Customer Segment:")
print(pivot_segment_sales)
```

```python
import matplotlib.pyplot as plt

# 1. Group the sales_df DataFrame by 'Year' and 'Month', and calculate the sum of 'Total_Sales'
monthly_sales = (
    sales_df
    .groupby(['Year', 'Month'])['Total_Sales']
    .sum()
    .reset_index()
)

print("Monthly Sales Data:")
print(monthly_sales)

# 2. Create a line chart for monthly sales trend
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Total_Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(monthly_sales['Month'].unique()) # Ensure all months are shown on x-axis
plt.grid(True)
plt.tight_layout()
plt.show()
```
